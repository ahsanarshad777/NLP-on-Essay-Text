```python
with open('essay.txt', 'r') as essay:
  essay_txt = essay.read()
```

```python
essay_txt = str(essay_txt)
```

```python
import nltk
nltk.download('punkt')
```

Sentence Tokenization

```python
sentences = nltk.sent_tokenize(essay_txt)
sentences
```

```python
for sent in sentences:
  print(sent)
  print()
```

Word Tokenization


```python
from nltk import word_tokenize
```

```python
words= []
for sentence in sentences:
  words.append(word_tokenize(sentence))
print(words)
print()
```

Lemmatization/Stemming

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lim = WordNetLemmatizer()
```

```python
for l in words:
  print(l)
```

```python
limted = []
num = 0
for l in words:
  word = words[num]
  for lk in word:
    limted.append(lim.lemmatize(lk))
  num = num+1
print(limted)
print()
```

```python
limted
```

```python
nltk.download('stopwords')
```

Removing Stopwords

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
without_stop_words = []

for word in limted:
  if word not in stop_words:
    without_stop_words.append(word)


print(without_stop_words)
print()
```

Bag of Words (BOW)


```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(without_stop_words)
feature_names = vectorizer.get_feature_names()
feature_names
```

```python
import pandas as pd
```

```python
feature = pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
```
